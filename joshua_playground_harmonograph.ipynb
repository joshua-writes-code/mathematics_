```python

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

#https://en.wikipedia.org/wiki/Harmonograph

```

```python
n = 1000000
t = np.logspace(np.log10(10), np.log10(500),n)

# parameters can be changed
A = [1, 2, 3, 5.01]
d = [0.004, 0.001, 0.02, .0015]
f = [3, 8, 7,5]

x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t))
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t + A[3]*np.sin(t*f[3])*np.exp(-d[1]*t))

#plot them
plt.plot(x,y, 'k', linewidth=0.1, color='black')
plt.axis('off')
#plt.show()

plt.savefig('plot_harmony')

```
